```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
data = ['the glass of milk',
        'the glass of wine',
        'the cup of tea',
        'i am a bad boy',
        'understanding the meaning of words',
        'i love movies']
```

```python
voc_size = 1000
```

```python
onehot_rep = [one_hot(words,voc_size)for words in data]
onehot_rep
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
#import numpy as np
```

```python
sent_length = 15
embedded_docs = pad_sequences(onehot_rep,padding='post',maxlen = sent_length)
embedded_docs
```

```python
dim = 15
```

```python
model = Sequential()
model.add(Embedding(voc_size,15,input_length = sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```

```python
def analysing_by_taking_input(user_input):
  list_inter = []
  one_hot_form = one_hot(user_input,voc_size)
  list_inter.append(one_hot_form)
  pad = pad_sequences(list_inter,padding='post',maxlen = sent_length)
  return pad
```

```python
user_input = input('Enter the sequence: ')
vector_form = analysing_by_taking_input(user_input)
print("vector_form : ",model.predict(vector_form))
```
